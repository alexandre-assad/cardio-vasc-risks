```python
from scripts.generic_methods import load_dataset
from icecream import ic

DATASET = load_dataset("../data/filtered/analysis.csv", ",")
```

```python
CATEGORICAL_FIELDS = [
    "ap_hi_status",
    "ap_lo_status",
    "cholesterol",
    "gluc",
    "broader_age_group",
    "in_hypertension",
    "is_healthy",
    "bmi_status",
    "gender",
    "smoke",
    "alco",
    "is_overweight",
    "is_underweight",
    "active"
]
```

```python
import matplotlib.pyplot as plt

for name in CATEGORICAL_FIELDS:
    DATASET.groupby([name, "cardio"]).size().unstack().plot(kind="bar")
    plt.xlabel(name)
    plt.ylabel("Count")
    plt.title(f"Number of Sick and Non-Sick Patients by {name}")

    plt.xticks(rotation=45)  # Rotate x-axis labels by 45 degrees

    plt.show()
```
